# Seismic Maximum August 4 2018

```python
# First, let's import the Python libraries we will need
import requests
import datetime
import matplotlib.pyplot as plt
import seaborn as sns # package for nice plotting defaults
sns.set()
import numpy as np
```

```python
USERNAME = 'OOIAPI-M8QQMKQSFYJ299'
TOKEN =  'TEMP-TOKEN-GJAGBOQSN4ZE8B'
```

```python
subsite = 'RS03ASHS'
node = 'MJ03B'
sensor = '07-TMPSFA301'
method = 'streamed'
stream = 'tmpsf_sample'
beginDT = '2017-08-04T10:01:01.000Z'
endDT = '2017-08-05T10:01:01.000Z'
```

```python
base_url = 'https://ooinet.oceanobservatories.org/api/m2m/12576/sensor/inv/'

data_request_url ='/'.join((base_url,subsite,node,sensor,method,stream))
params = {
    'beginDT':beginDT,
    'endDT':endDT,
    'limit':1000,   
}

# Grab the data
r = requests.get(data_request_url, params=params, auth=(USERNAME, TOKEN))
data = r.json()
```

```python
data[0]
```

```python
import netCDF4 as nc
```

# Temperature Maximums, August 4 2018

```python
time = []
temperature01 = []

for i in range(len(data)):
    time.append(nc.num2date(data[i]['time'],'seconds since 1900-01-01').replace(microsecond=0))
    temperature01.append(data[i]['temperature01'])
```

```python
plt.plot(time, temperature01, marker=".", markersize=1, linestyle=None)
plt.ylabel('Temperature $^\circ$C')
plt.xlabel('Time')
plt.xticks(rotation=30)
plt.tight_layout()
plt.show()
```
